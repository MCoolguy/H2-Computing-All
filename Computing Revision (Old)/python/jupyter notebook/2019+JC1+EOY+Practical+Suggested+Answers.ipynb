```python
#Question 1
import random

side1 = [0.0]*100000
side2 = [0.0]*100000
side3 = [0.0]*100000
triangle_formed = [0]*100000

for trial in range(100000):
       b1 = random.random()
       b2 = random.random()

       s1 = min(b1,b2)
       s2 = abs(b1-b2)
       s3 = 1-(s1+s2)

       side1[trial] = (round(s1,3))
       side2[trial] = (round(s2,3))
       side3[trial] = (round(s3,3))
       
       diff = max(s1,s2,s3)-(1-(max(s1,s2,s3)))
       
       if diff < 0:
              triangle_formed[trial] = 1
       else:
              triangle_formed[trial] = 0

print("Side1\tSide2\tSide3\tTriangle_formed")

for i in range(10):
       print(str(side1[i])+"\t"+str(side2[i])+"\t"+str(side3[i])+"\t"+str(triangle_formed[i]))

print()
total = sum(triangle_formed,0)

print(round(total/100000,2))
```

```python
#Question 2
def MERGESORT(Scores):
    if len(Scores)>1:
        mid = len(Scores)//2
        lefthalf = Scores[0:mid]  #copy of subarray
        righthalf = Scores[mid:len(Scores)] #copy of subarray

        #missing codes
        #=============
        MERGESORT(lefthalf)     #caller function proceeds to next line when len == 1, recursive function returns None
        MERGESORT(righthalf)    #caller function proceeds to next line when len == 1, recursive function returns None
        #==============
        
        i=0 #pointer for lefthalf
        j=0 #pointer for righthalf
        k=0 #pointer for original Scores 
        while i < len(lefthalf) and j < len(righthalf):
            if lefthalf[i] <= righthalf[j]:
                Scores[k]=lefthalf[i] 
                i=i+1
            else:
                Scores[k]=righthalf[j]
                j=j+1
            k=k+1

        while i < len(lefthalf):
            #missing codes
            #=============
            Scores[k] = lefthalf[i]
            i = i+1
            k = k+1

        while j < len(righthalf):
            #missing codes
            #=============
            Scores[k] = righthalf[j]
            j = j+1
            k = k+1

def DISPLAY(lst):
    print(" ".join(str(item) for item in lst))    

Scores = [72,45,57,76,29,40,42,77,64,91,24,92,73,78,62,27,26,67,34,20,90,75,89,98,41,37,58]

print("Unsorted Scores")
DISPLAY(Scores)
MERGESORT(Scores)
print()
print("Sorted Scores")
DISPLAY(Scores)
```
