```python
# Task 3.1

class HashTable:
    def __init__(self):
        self.data = [None] * 31
    
    def hash(self, membership_num):
        return membership_num % 31
    
    def display(self):
        """Display the contents of the hash table"""
        for item in self.data:
            if item is not None:
                id_, name = item
                print(f"{id_}: {name}")
    
    def add(self, id_, name):
        index = self.hash(id_)
        if self.data[index] is not None:
            print(f"{id_}: Address is occupied")
        else:
            self.data[index] = (id_, name)
```

```python
# Task 3.3

class HashTable:
    def __init__(self):
        self.data = [None] * 31
    
    def hash(self, membershipNo):
        return membershipNo % 31
    
    def rehash(self, membershipNo):
        return 2**membershipNo % 31
    
    def display(self):
        """Display the contents of the hash table"""
        for i, item in enumerate(self.data):
            if item is not None:
                membershipNo, name = item
                print(f"[{i}] {membershipNo}: {name}")
    
    def add(self, membershipNo, name):
        index = self.hash(membershipNo)
        if self.data[index] is not None:
            index = self.rehash(membershipNo)
            if self.data[index] is not None:
                print(f"{membershipNo}: Address is occupied")
            else:
                self.data[index] = (membershipNo, name)
        else:
            self.data[index] = (membershipNo, name)
    
    def search(self, membershipNo):
        index = self.hash(membershipNo)
        stored_membershipNo, name = self.data[index]
        if stored_membershipNo == membershipNo:
            return name
        else:
            index = self.rehash(membershipNo)
            stored_membershipNo, name = self.data[index]
            if stored_membershipNo == membershipNo:
                return name
            else:
                print(f"Member {membershipNo} not found")



ht = HashTable()

with open('IDs.txt', 'r') as f:
    for line in f:
        membershipNo, name = line.strip().split(',')
        membershipNo = int(membershipNo)
        ht.add(membershipNo, name)

ht.display()
```
