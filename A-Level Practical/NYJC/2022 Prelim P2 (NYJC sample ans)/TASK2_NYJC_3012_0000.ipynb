```python
# Task 2.1

import datetime

class LibraryItem:
    def __init__(self, item_id: str, title: str):
        self.item_id = item_id
        self.title = title
        self.loaned_by = ''
        self.due_date = ''
    
    def is_on_loan(self) -> bool:
        return not (self.loaned_by == '')
    
    def return_item(self):
        self.loaned_by == ''

    def loan_to(self, user_id) -> bool:
        if self.loaned_by != '':
            return False
        self.loaned_by = user_id
        newdate = datetime.date.today() + datetime.timedelta(weeks=3)
        self.due_date = newdate.strftime('%d-%m-%Y')
        return True
    
    def print_details(self):
        print('Title:', self.title)
        if self.loaned_by != '':
            print('Loaned by:', self.loaned_by)
            print('Due date:', self.due_date)
```

```python
# Task 2.2

class Book(LibraryItem):
    def __init__(self, item_id: str, title: str, author: str, category: str):
        super().__init__(item_id, title)
        self.author = author
        self.category = category
        self.reserved_by = ''
    
    def is_reserved(self):
        return not (self.reserved_by == '')
    
    def release(self):
        self.reserved_by = ''
    
    def reserve_for(self, user_id):
        if self.reserved_by != '':
            return False
        self.reserved_by = user_id
        return True
    
    def print_details(self):
        print('Title:', self.title)
        print('Author:', self.author)
        print('Category:', self.category)
        if self.loaned_by != '':
            print('Loaned by:', self.loaned_by)
            print('Due date:', self.due_date)



class CompactDisc(LibraryItem):
    def __init__(self, item_id: str, title: str, artist: str, genre: str):
        super().__init__(item_id, title)
        self.artist = artist
        self.genre = genre

    def print_details(self):
        print('Title:', self.title)
        print('Artist:', self.artist)
        print('Genre:', self.genre)
        if self.loaned_by != '':
            print('Loaned by:', self.loaned_by)
            print('Due date:', self.due_date)
```

```python
# Task 2.3

import csv

libraryitems = []
with open('libraryitems.txt', 'r') as f:
    for row in csv.reader(f):
        if row[0].startswith('B'):
            item_id, title, author, category = row
            item = Book(item_id, title, author, category)
        elif row[0].startswith('C'):
            item_id, title, artist, genre = row
            item = CompactDisc(item_id, title, artist, genre)
        item.print_details()
        print()
        libraryitems.append(item)
```

```python
# Task 2.4

for item in libraryitems:
    item.loan_to('S1111111H')

for item in libraryitems:
    if isinstance(item, Book) and item.loaned_by == 'S1111111H':
        item.reserve_for('S1222222D')

for item in libraryitems:
    item.print_details()
    print()
```
