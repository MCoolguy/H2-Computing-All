```python
#Task 3.1
class DataStructure():
    def __init__(self):
        self.linked_lst = []
        self.head_pointer = -1
        self.tail_pointer = -1
    
    def add(self, value):
        if self.head_pointer == -1:
            self.head_pointer = 0
        node = Node(value, -1) #-1 is a null value, since the node added is at the end of the linked list
        self.linked_lst.append(node)
        if self.tail_pointer != -1:
            self.linked_lst[self.tail_pointer].set_pointer(self.linked_lst.index(node))
        self.tail_pointer = self.linked_lst.index(node)
        return
    
    def display(self):
        if self.head_pointer == -1:
            print("Linked list is empty")
            return
        
        print("{0:8}{1:15}".format("Order", "Content"))
        n = 1 #order number
        
        temp_node = self.linked_lst[self.head_pointer]
        pointer = temp_node.get_pointer()
        while pointer != -1:
            print("{0:8}{1:15}".format(str(n), str(temp_node.get_data())))
            n += 1
            temp_node = self.linked_lst[temp_node.get_pointer()] #next item in the linked list
            pointer = temp_node.get_pointer()
            
        print("{0:8}{1:15}".format(str(n), str(temp_node.get_data())))
        
        
class Node():
    def __init__(self, Data, Pointer):
        self.data = Data
        self.pointer = Pointer
    
    def set_pointer(self, new_pointer):
        self.pointer = new_pointer
    
    def get_pointer(self):
        return self.pointer
    
    def set_data(self, Data):
        self.data = Data
        
    def get_data(self):
        return self.data
    
    
```

```python
#Task 3.2
class Stack(DataStructure):
    def __init__(self):
        super().__init__()
    
    def insert(self, value):
        super().add(value)
    
    def delete(self):
        node = self.linked_lst[self.tail_pointer]
        self.tail_pointer -= 1
        if self.tail_pointer == -1: #last item in the list:
            #reinitialise
            self.tail_pointer = -1
            self.linked_lst = []
        
        print(node.get_data())
    
    def display(self):
        if self.head_pointer == -1:
            print("Linked list is empty")
            return
        
        print("{0:8}{1:15}".format("Order", "Content"))
        n = 1 #order number
        
        lst = [] #to store the items in reversed order
        temp_node = self.linked_lst[self.head_pointer]
        pointer = temp_node.get_pointer()
        while pointer != -1:
            data, pointer = temp_node.get_data(), temp_node.get_pointer()
            n += 1
            lst = [data] + lst
            temp_node = self.linked_lst[pointer] #next item in the linked list
            if pointer == self.tail_pointer: #it is at last element:
                data = temp_node.get_data()
                lst = [data] + lst
                break
                
        for item in lst:
            print("{0:8}{1:15}".format(str(n), item))
            n -= 1
```

```python
#Task 3.3
class Queue(DataStructure):
    def __init__(self):
        super().__init__()
        
    def insert(self, value):
        super().add(value)
    
    def delete(self):
        node = self.linked_lst[self.head_pointer]
        self.head_pointer = node.get_pointer()
        if self.head_pointer == -1: #last item in the list:
            #reinitialise
            self.tail_pointer = -1
            self.linked_lst = []
        
        print(node.get_data())
        
    def display(self):
        super().display()
```

```python
#Task 3.4
stack = Stack()
queue = Queue()

stack_file = open("TASK3stack.txt", 'r')
queue_file = open("TASK3queue.txt", 'r')

#reading and adding values from file to objects
for line in stack_file:
    value = line.strip()
    stack.insert(value)

for line in queue_file:
    value = line.strip()
    queue.insert(value)

#output content
print("stack:")
stack.display()
print()
print("queue:")
queue.display()
print()

#removal of 2 items
print("Removing Stack Items")
for count in range(2):
    stack.delete()
print()

print("Removing Queue Items")
for count in range(2):
    queue.delete()
print()

#output content
print("stack:")
stack.display()
print()
print("queue:")
queue.display()
print()


#closing of files
stack_file.close()
queue_file.close()
```
