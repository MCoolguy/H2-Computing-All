```python
#Task 1.2
def Task2_2():
    userinput = input("Enter a letter: ")
    base = input("Enter a number base: ")
    while int(base)<10 or int(base)>15:
        base = input("Enter a number base: ")
    
    denaryvalue = ord(userinput)
    numberbase = ""
    quotient = denaryvalue
    
    while quotient!=0:
        remainder = quotient % int(base)
        quotient = quotient // int(base)
        numberbase += str(remainder)
    
    numberbase = numberbase[::-1]
    
    print("Letter".ljust(20) + userinput)
    print("Denary".ljust(20) + str(denaryvalue))
    print(("Number Base "+base).ljust(20)+ numberbase)
   

    
for i in range(2):
    Task2_2()

        
```
