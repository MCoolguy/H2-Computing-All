```python
#Task 3.1
class DataStructure:
    def __init__(self):
        self.LL = [""for i in range(12)]
        self.head = -1
        self.tail = 0
    
    def display(self):
        pass
        
class Stack(DataStructure):
    def __init__(self):
        super().__init__()
        
    def insert(self,value):
        if self.head ==-1:
            self.head = 0
            self.LL[self.head]  = value
        else:
            self.tail+=1
            self.LL[self.tail] = value
            
    
    def delete(self):
        temp = self.LL[self.tail]
        self.tail -=1
        return temp
    
    def display(self):
        print("Stack contains: " + str(self.LL[self.head:self.tail+1]))
    

class Queue(DataStructure):
    def __init__(self):
        super().__init__()
    
    def insert(self,value):
        if self.head ==-1:
            self.head = 0
            self.LL[self.head] = value
        else:
            self.tail +=1
            self.LL[self.tail] = value
    
    def delete(self):
        temp = self.LL[self.head] 
        self.head +=1
        return temp
    
    def display(self):
        print("Queue contains: " + str(self.LL[self.head:self.tail+1]))
        

        
##MAIN##
stack = Stack()
stackfile = open("TASK3stack.txt","r")
for line in stackfile:
    stack.insert(line.strip())
stackfile.close()
    
queue = Queue()
queuefile = open("TASK3queue.txt","r")
for line in queuefile:
    queue.insert(line.strip())
queuefile.close()

stack.display()
queue.display()

print("Stack removes: " + str(stack.delete()))
print("Stack removes: " + str(stack.delete()))
print("Queue removes: " + str(queue.delete()))
print("Queue removes: " + str(queue.delete()))
stack.display()
queue.display()
```
