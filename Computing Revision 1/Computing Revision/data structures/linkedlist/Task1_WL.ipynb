```python
class Node:   #suitable record structure, attributes, methods
    def __init__(self): 
        self.record = ['','',''] #record stores the name, contact and birthday 
        self.pointer = -1
        
    def getRecord(self):
        return self.record
    
    def setRecord(self, record):
        self.record = record
        
    def getPointer(self):
        return self.pointer
    
    def setPointer(self, newPtr):
        self.pointer = newPtr

class LinkedList:
    def __init__(self, size):  #array of nodes, linked unused nodes, start and nextfree
        self.ArrayOfNodes = [Node() for i in range(size)]
        
        #link the unused nodes
        for i in range(size-1):
            self.ArrayOfNodes[i].setPointer(i+1)
            
        self.start = -1
        self.nextfree = 0   

    def insert(self, record):  
        self.ArrayOfNodes[self.nextfree].setRecord(record)  
        temp = self.ArrayOfNodes[self.nextfree].getPointer() 
        
        if self.start == -1:   #empty linked list   
            self.start = self.nextfree
            self.ArrayOfNodes[self.nextfree].setPointer(-1)
        else:
            if self.ArrayOfNodes[self.start].getRecord()[2] > record[2]:   #insert as first item  
                self.ArrayOfNodes[self.nextfree].setPointer(self.start)    #points to previous start node
                self.start = self.nextfree
            else:  #new item is not inserted as first item
                prev = -1
                curr = self.start

                while True:
                    if self.ArrayOfNodes[curr].getRecord()[2] > record[2]: #position found
                        self.ArrayOfNodes[self.nextfree].setPointer(curr)
                        self.ArrayOfNodes[prev].setPointer(self.nextfree) 
                        break
                        
                    if self.ArrayOfNodes[curr].getPointer() == -1: #reached end of linked list
                        self.ArrayOfNodes[curr].setPointer(self.nextfree) 
                        self.ArrayOfNodes[self.nextfree].setPointer(-1) 
                        break
                        
                    prev = curr
                    curr =  self.ArrayOfNodes[curr].getPointer()
         
        self.nextfree = temp

    def display(self):  
        print("{:<10}|{:<20}|{:<10}|{:<15}|{:<10}".format("Index","Name","Contact","Birthday","Pointer"))  
        for i in range(len(self.ArrayOfNodes)):  
            print("{:<10}|{:<20}|{:<10}|{:<15}|{:<10}".format(i,str(self.ArrayOfNodes[i].getRecord()[0]),\
                                                              str(self.ArrayOfNodes[i].getRecord()[1]),\
                                                              str(self.ArrayOfNodes[i].getRecord()[2]),\
                                                              str(self.ArrayOfNodes[i].getPointer())))
        print("start",self.start)          
        print("nextfree",self.nextfree)
        
    def traverse(self):
        curr = self.start
        while True:
            if curr == -1:
                break
                
            print(self.ArrayOfNodes[curr].getRecord()[2])
            curr = self.ArrayOfNodes[curr].getPointer()
                
            
#Main Program
info = list()
file = open("birthdayFile.txt",'r')  

for line in file:
    info.append(line.strip())   
    
file.close()
linkedlist1 = LinkedList(20)          
linkedlist1.display()

for i in range(len(info)):    
    if i%3 == 0:
        name = info[i]
    elif i%3 == 1:
        contact = info[i]
    elif i%3 == 2:
        birthday = info[i]
        linkedlist1.insert([name, contact, birthday])   
        
linkedlist1.display()
print()       
linkedlist1.traverse()
```
